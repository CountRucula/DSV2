```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

## Aufgabe 1

```python
z = np.array([0.64, 0.01, 0.02, 0.31, 0.95, 0.90, 0.73, 0.66, 0.62, 0.21])

print(f'mean = {np.mean(z):.2f}')
print(f'power = {np.mean(z**2):.2f}')
print(f'Var = {np.var(z):.2f}')
print(f'std = {np.std(z):.2f}')
```

## Aufgabe 2

```python
ps = np.array([1,2,3,4,5,6,5,4,3,2,1], dtype=float)/36.0
s = np.arange(start=2, stop=12+1, dtype=float)

mean = np.sum(ps*s)
print(mean)

quad_maen = np.sum(ps*(s**2))
print(quad_maen)

var = quad_maen - mean**2
print(var)
```

```python
x = np.linspace(2, 12, 100)
gauss = norm.pdf(x, mean, np.sqrt(var))

plt.figure()
plt.plot(x,gauss, s, ps)
plt.grid(True)
plt.legend(('Gaussche Approximation', 'Exakte Verteilung'))

plt.xlabel('S')
plt.ylabel(r'$P_s(S)$')
```

## Aufgabe 3

```python
M = 1.0
N = 16
var = 1.0
std = np.sqrt(var)
mu = 0

z = np.linspace(M-3*std, M+3*std, 1000)

x = norm.pdf(z, mu+M, var)
y = norm.pdf(z, mu+M, var/N)

plt.figure()
plt.plot(z, x, z, y)
plt.legend(("Ohne Mittelung", "Mittelung ber 16 Samples"))
plt.xlabel('Z')
plt.ylabel(r'$P_Z(z)$')
plt.grid(True)
plt.show()
```
